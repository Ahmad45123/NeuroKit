```python
import neurokit2 as nk
from neurokit2.rsp.rsp_rvt import rsp_rvt
import matplotlib.pyplot as plt
```

```python
sums = []
for respiratory_rate in range(1,20):
    print(respiratory_rate)
    rsp = nk.rsp_simulate(duration=90, respiratory_rate=respiratory_rate, random_state=41)
    rsp = nk.rsp_clean(rsp)
    rvt = rsp_rvt(rsp, show=True, silent=False)
    plt.plot(rsp)
    sums.append(sum(rvt))
    assert len(rsp) == len(rvt)
    assert min(rvt) >= 0
```
