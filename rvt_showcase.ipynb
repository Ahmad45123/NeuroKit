```python
import neurokit2 as nk
import matplotlib.pyplot as plt

for seed in range(6):
    print(f"The current seed is {seed}.")
    rsp10 = nk.rsp_simulate(duration=60, sampling_rate=1000,respiratory_rate=10, random_state=seed)
    for method in ["harrison", "power", "birn"]:
        print(method)
        rvt10 = nk.rsp_rvt(rsp10, method=method, show=True)
        plt.show()
```
